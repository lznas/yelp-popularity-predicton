What observable characteristics of restaurants in Philidelphia — including average rating, review volume, price level, cuisine type, geographic location, and sentiment patterns in customer reviews — are associated with higher popularity? Which of these features provide the strongest predictive signals for identifying highly popular restaurants? ## Background and Prior Work

The restaurant industry is inherently volatile, and as time passes, maintaining a restaurant has become increasingly difficult and costly due to high inflation, surging labor costs, and shifting consumer habits. In fact, in 2025, 42% of restaurants were reported to be unprofitable.<sup>[1]</sup> Compare this to the estimated 60–70% of profitable establishments just ten years ago.<sup>[2]</sup> If a restaurant could accurately predict profitability within a specific geographical location or the viability of incorporating a certain cuisine, it could more reliably guarantee meeting an annual budget and increase its chances of longevity. Furthermore, existing restaurants can analyze the differences between high-performing and low-performing establishments to adopt best practices and improve overall performance.

Historically, the 1990s present a stark contrast: prior to the dominance of digital media, restaurant survival was lower in the first year due to significant barriers in spreading information and attracting customers. At that time, roughly 70% of restaurants survived their first year, compared to 83% today.<sup>[4]</sup> However, while first-year survival rates have generally increased, modern day restaurants face a new set of obstacles. Establishments now witness steeper drop-off rates between years one and five due to margin erosion, a phenomenon where business costs rise faster than menu prices. Ultimately, this has resulted in 10% fewer establishments surviving a five year period today than in the 1990s.<sup>[2][4]</sup>

Market conditions fluctuate tremendously even from year to year. Business trends in 2024 differed significantly from those in 2025, with 60% of restaurants reporting deteriorated conditions by the end of 2024.<sup>[3]</sup> The same volatility applies to profit margins between 2025 and 2026; margins in 2026 are predicted to grow primarily from the consumption of affluent households, despite a general pull-back in spending from middle- and lower-class households.<sup>[3]</sup> External factors like tariffs on imported food and beverage items have also posed challenges for two-thirds of operators. Average food costs are now more than 35% above pre-pandemic levels, according to the U.S. Bureau of Labor Statistics. In 2025, 82% of operators reported higher average food costs, and 68% noted that tariffs drove those costs higher.<sup>[3]</sup> 

Despite this economic pressure, some locations have thrived; for example, in 2025, Chili’s reported a 13% increase in foot traffic and a 23% increase in sales.<sup>[3]</sup> Consequently, it is becoming vital to build predictive models for restaurant performance. By analyzing the practices of profitable restaurants like Chili’s—considering variables such as average ratings, review volume, price level, cuisine type, geographic location, and sentiment patterns—operators can better mitigate financial risks and seize profit opportunities.

The role of COVID-19 in restaurant longevity research cannot be overlooked. The pandemic acted as a "forced laboratory," intensifying the search for variables that predict success and longevity. Before the pandemic, failure was largely attributed to poor location or general service quality issues that led to a lack of capital. At that time, the average lifespan of an establishment was approximately 4.5 years. Post-COVID, this figure stands significantly lower, at 1 to 3 years for those failing to adapt.<sup>[5]</sup> A key quantifier in this change is the heavy dependency on delivery and digital integration as a primary survival variable.

Current research indicates that longevity is no longer found through traditional management alone, but through the integration of multimodal data. Studies from late 2025 and early 2026, such as those from Hong Kong Polytechnic University, have introduced "Text–Image Fusion" models that analyze customer-generated content to predict failure with high precision.<sup>[6]</sup> Unlike older models that relied solely on financial spreadsheets, these AI-driven systems analyze the aesthetic quality of food photos, sentiment patterns in reviews, and even meteorological data to forecast demand. Today, a restaurant's ability to survive the five-year mark is heavily dependent on how well it optimizes for delivery only zones and utilizes dynamic pricing to capture value-conscious diners. These predictive models allow operators to transition from reactive survival to proactive growth by identifying the exact cuisine and geographic variables that align with modern consumer trends.

### Yelp Open Dataset - Businesses Queried to Philidelphia
The Yelp Open Dataset's businesses dataset queried to Philidelphia contains resturant specific data, where each row is represented by an unique identifier for each individual business (identified by business_id). The most important quantatiative metric we intend to look at is stars, rated on a 1.0-5.0 scale that is incremented by 0.5. This means even small changes may be a reprentation of a meaningful difference. Another key variable is review_count, or the total number of reviews. This will be a proxy for visibility and customer engagement to answer whether resturants with higher number of reviews signifying a more established business, with more traffic. Variables such as categories (cuisine types such as Italian, Mexican, or Chinese), attributes (describing features of the resturant), and geographic variables like latitude and longitude (measured in degrees), which allow for spatial analysis across neighborhoods in Philadelphia.

There are several limitations to keep in mind when working with this dataset. As this dataset only includes business-level information and aggregated ratings, we cannot use this datset to under how customer behavior, or calculate actual foot traffic. Additionally, it is importatn to keep in mind that review_count is influenced by not only its popularity but also how long a resturant has been on Yelp. There may also be a self-selection bias where resturants with a stronger online presence may accumulate more reviews and its important to note that while these businesses are real, this data set was currated by Yelp who may favor resturants that are more popular affectiing the dataset we are analyzing. While, review_count is likely the most observable sign of popularity, alongside stars to create a composite popularity measure. Cusine type and location can also be investiagted to explain popularity. However, the model built from these variables should be interpreted with caution as we are using proxies of poularity instead to using direct measures of business success.

Although some variables such as `address`, `attributes`, `categories` and `hours` contain missing data, these fields are not a central part of or modeling objective. Instead, we retained all businesses with complete core variables (e.g., `business_id`, `stars`, `review_count`, geographic coordinates).

To check if the missing data is systematic, we will compare the values with and without missing values to see if there is a significant difference. 

```python
# Comapre stars and review counts with and without `hours`
yelp_ph_clean.groupby(
    yelp_ph_clean["hours"].isna()
)[["stars", "review_count"]].mean()
```

While average star ratings were similar between groups, businesses missing `hours` exhibited substantially lower average review counts (19 vs 75). This suggests that missingness may be systematic such as due to lower engagement rather than occurring completely at random.

We will do the same for `attributes`.

```python
# Comapre stars and review counts with and without `attributes`
yelp_ph_clean.groupby(
    yelp_ph_clean["attributes"].isna()
)[["stars", "review_count"]].mean()
```

Businesses missing `attributes` exhibit substantially lower average review counts (18 vs 68) and slightly lower average ratings (3.27 vs 3.65), suggesting that missingness is systematic and associated with lower engagement. This indicates that incomplete business profiles may correlate with reduced popularity.

A.2 Collection bias: Have we considered sources of bias that could be introduced during data collection and survey design and taken steps to mitigate those?

Yelp data may introduce selection bias because reviewers are self-selected and may not represent the broader population of restaurant patrons. Certain neighborhoods, cuisines, or price levels may also receive disproportionate attention. As a result, our measure of “popularity” may reflect online engagement rather than true customer volume, and we will interpret findings accordingly.

A.3 Limit PII exposure: Have we considered ways to minimize exposure of personally identifiable information (PII) for example through anonymization or not collecting information that isn't relevant for analysis?

Our predictive model could unintentionally favor restaurants in higher-income or high-traffic areas, reinforcing existing socioeconomic disparities. While we are not collecting protected attributes such as race or gender, we will examine whether model performance varies across neighborhoods or cuisine types to assess potential structural bias.

D. Modeling

D.1 Proxy Discrimination: The primary risk here is using geographic location as a proxy for socioeconomic status or race. In Philadelphia, ZIP codes can reflect historical redlining or neighborhood wealth; if the model relies too heavily on location to predict "popularity," it may unfairly penalize businesses in underserved areas. We must ensure the model focuses on service-related features rather than location-based demographics that correlate with protected groups.

D.2 Disparate Impact: The model might achieve high accuracy overall but fail for specific cuisine types or niche markets. For instance, if the dataset is dominated by American fast food, the model’s "popularity predictors" might not accurately apply to authentic, family-owned ethnic restaurants in neighborhoods like Chinatown, South Philadelphia, or West Philadelphia. We should audit error rates across different categories to ensure the model isn't systematically "underrating" the potential of diverse businesses.

D.3 Metric Selection: Optimizing for average rating alone can be misleading, as a restaurant with five reviews is not necessarily more "popular" than one with 1,000 reviews and a 4-star average. We should use a weighted metric or a combination of review volume and sentiment density to capture a truer sense of popularity. This prevents the model from being skewed by small sample sizes or "review bombing" behavior.

D.5 Communicate Limitations: We must inform stakeholders that Yelp data is a "snapshot" and subject to selection bias, as not everyone leaves reviews. The model reflects the opinions of Yelp users—who may skew toward a specific demographic—rather than the entire Philadelphia population. Additionally, the model cannot account for external factors like sudden "TikTok trends" or road construction that might temporarily impact a restaurant’s foot traffic.

E. Deployment

E.1 & E.3 Monitoring and Rollback: Since popularity trends shift seasonally, we need to monitor the model for concept drift to ensure its predictions remain relevant. Because we are using Git for version control, we have a safety net: if a model update begins producing biased or nonsensical recommendations for certain Philadelphia neighborhoods, we can immediately revert to a previous, stable version of the code.

E.4 Unintended Use: While this tool is meant to help small business owners understand growth, it could be unintentionally used by real estate speculators or predatory lenders to identify areas for gentrification. We must clarify that the model is designed for operational improvement—helping a restaurant refine its menu or service—not as a tool to exclude or exploit businesses in "low-popularity" predicted zones.
