## Hypothesis


- Customer churn is expected to vary based on tenure, service configuration, billing changes, monthly usage, gender, and other demographic information. In particular, customers with shorter tenure, higher monthly charges, higher billing charges, without partners or dependents, and with service configurations such as multiple phone lines, and certain internet service types are hypothesized to have higher churn rates, while customers with longer tenure and additional service investments such as online security are expected to churn less frequently. 

- Our thinking behind this is that customers don't leave randomly. They usually leave a service after experiencing things like rising costs, low perceived value, flexibility to switch, and limited long term commitment. 

## Data

**Ideal Data Set**

Our ideal dataset would be a relational database containing anonymized customer IDs (to align with privacy standards) and consisting of one table with static customer attributes (eg, demographic information) and one or more tables containing information about records, updated monthly. Ideally, we would have a table that is time-indexed and lists monthly usage information, as well as another table with billing information, due to the assumption that these two variables heavily impact churn rates. 
Significant variables would include our target variable (churned or retained for the next month), account and contract information (eg, contract type, monthly charges or changes, customer longevity), service usage (eg, type of service, service usage), customer experience (eg, number of customer support interactions, service interruptions, customer service ratings), and demographic information (eg, age range, gender, household composition, geographic region). 
Creating an accurate prediction model would require a large amount of individual customer data, ideally at least 10,000, for equal representation of sources of churning. We would have access to 12-24 months' worth of observations per customer to analyze trends that may predict churning or retention. Furthermore, ensuring equal representation of customers who continue to use the service alongside customers who leave within the dataset will ensure a more accurate, robust prediction model.
In order to collect the most accurate data, we would want to get this information from the internal systems of a telecommunications provider. For instance, account and billing data would be collected from the customer relationship management and billing branches of the company.  Service usage data could be collected from the network or usage logs. Appropriate consent, anonymization, and additional safeguards will ensure that there is responsible data handling and privacy considerations. Through customer support and ticket systems, we could collect and organize customer interactions that are likely to correlate with churning. Due to ethical concerns, we would incentivise customers to share their demographic information and satisfaction data and only use data collected through voluntary customer surveys.
The finalized training dataset would be created through aggregation of the previously collected and organized tables, focusing on consumer behavior prior to churning, and exported to a structured format for analysis and modeling.
