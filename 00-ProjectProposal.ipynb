## Research Question

What characteristics of a customer’s account, service usage, and background are associated with leaving the service, and which of these factors provide the most useful signals for identifying at-risk customers? To what extent can these features be used to build a model that anticipates customer departures and separates churned customers from retained ones?



## Background and Prior Work

How do you predict profitability of a customer? One mode of increasing profits of a company is undoubtedly increasing customer loyalty, but customer loyalty depends not only on the quality of the product or other assisted services. What if there were some other variables that could measure the reliability of a customer, such as the other services each customer has and their respective usage, and their general account information/demographics in which we can predict whether or not a customer is likely to be loyal and therefore provide long-term profit? Conversely, can you predict what customers are likely to become inactive, or in other words “churned” out from these same variables in order to potentially maintain these at-risk customers?

It is no secret that it is much more expensive to attract new customers than it is to keep existing ones. An existing customer already has trust in the product and its providers, whilst a new one needs convincing, through advertising, multiple touchpoints from sales representatives, and other means. In fact, research from Frederick Reichheld shows that increasing customer retention by just 5% can increase company profits from 25% up to 95%.<sup>1</sup> Furthermore, costs of gaining a new customer can be anywhere from five to twenty five times more expensive than maintaining an existing one.<sup>1</sup> This alone highlights the importance of maximizing customer loyalty and henceforth predicting which customers are most likely to maintain high levels of loyalty.

Where profit is concerned, research will ultimately follow. That is, there have been significant advances in the field since the 1980s, when it was first identified as a high-profit business strategy. Today companies utilize precise and reliable predictive models to predict customer retention. Such research highlights the extreme predictive power of usage (frequency, velocity, feature depth) specifically. By tracking usage current models can detect flow fades in engagement which are the strongest and most intervention worthy signals of impending departure. Other variables with high predictive power are financial indicators such as payment failures and monthly charges as these display a customer's ability and intent to pay. Furthermore, support variables such as complaint frequency and issue resolution time act as high risk indicators.

One tremendous benefit to identifying at risk customers is that they can be targeted with increased advocacy in order to retain said customer. Once these at-risk customers are identified, campaigns can be triggered that involve addressing raised or implied concerns, providing direct lines of communication, or offering perks or upgrades<sup>2</sup> in prophylactic effort. In fact, further research delves into whether or not we can predict if these at-risk customers are recoverable or not, showing that customers that leave due to price sensitivity are statistically more likely to be won back than those that left due to poor service quality.<sup>3</sup> So not only can variables such as service usage be used to identify indicators of customer departure, but they can provide roadmaps to recovery.

Impressively, there have been advances in refining predictive models for even highly volatile fields, such as the music industry, where customer churn is unpredictable and thus poses an increased threat to company sustainability. One such example is the usage of learning techniques such as XGBoost and Random Forest that outperform traditional models by processing more complex features such as payment method shifts<sup>4</sup>, and thus more precise interventions can be implemented before these at-risk listeners discontinue their subscriptions.
It is thus our intent here to confirm for ourselves that these variables in fact do provide high predictability, as well as possibly identifying if there are other variables with predictive power that may have been overlooked. Existing models rely heavily on financial and usage data, but perhaps more human elements can provide predictability such as analysis of support chats.

## Hypothesis


- Customer churn is expected to vary based on tenure, service configuration, billing changes, monthly usage, gender, and other demographic information. In particular, customers with shorter tenure, higher monthly charges, higher billing charges, without partners or dependents, and with service configurations such as multiple phone lines, and certain internet service types are hypothesized to have higher churn rates, while customers with longer tenure and additional service investments such as online security are expected to churn less frequently. 

- Our thinking behind this is that customers don't leave randomly. They usually leave a service after experiencing things like rising costs, low perceived value, flexibility to switch, and limited long term commitment. 

## Data

**Ideal Data Set**

Our ideal dataset would be a relational database containing anonymized customer IDs (to align with privacy standards) and consisting of one table with static customer attributes (eg, demographic information) and one or more tables containing information about records, updated monthly. Ideally, we would have a table that is time-indexed and lists monthly usage information, as well as another table with billing information, due to the assumption that these two variables heavily impact churn rates. 
Significant variables would include our target variable (churned or retained for the next month), account and contract information (eg, contract type, monthly charges or changes, customer longevity), service usage (eg, type of service, service usage), customer experience (eg, number of customer support interactions, service interruptions, customer service ratings), and demographic information (eg, age range, gender, household composition, geographic region). 
Creating an accurate prediction model would require a large amount of individual customer data, ideally at least 10,000, for equal representation of sources of churning. We would have access to 12-24 months' worth of observations per customer to analyze trends that may predict churning or retention. Furthermore, ensuring equal representation of customers who continue to use the service alongside customers who leave within the dataset will ensure a more accurate, robust prediction model.
In order to collect the most accurate data, we would want to get this information from the internal systems of a telecommunications provider. For instance, account and billing data would be collected from the customer relationship management and billing branches of the company.  Service usage data could be collected from the network or usage logs. Appropriate consent, anonymization, and additional safeguards will ensure that there is responsible data handling and privacy considerations. Through customer support and ticket systems, we could collect and organize customer interactions that are likely to correlate with churning. Due to ethical concerns, we would incentivise customers to share their demographic information and satisfaction data and only use data collected through voluntary customer surveys.
The finalized training dataset would be created through aggregation of the previously collected and organized tables, focusing on consumer behavior prior to churning, and exported to a structured format for analysis and modeling.

### D. Modeling
 - [ ] **D.1 Proxy discrimination**: Have we ensured that the model does not rely on variables or proxies for variables that are unfairly discriminatory?

> The dataset has columns for things like Gender, SeniorCitizen, Partner, and Dependents. Even though these might help the model predict churn, using them directly is risky. For example, if the model figures out that older people (Senior Citizens) are more likely to leave and we treat them differently because of it, that could be considered age discrimination. We have to be careful about which features we actually feed into the model.
D2. We need to make sure the model doesn't just work well for the "average" customer but fails for specific groups. For instance, if the training data is mostly young people, the model might not learn the patterns for senior citizens very well. We should check the error rates (like false negatives) specifically for different demographics to ensure we aren't systematically missing churn signals for one group over another.

 - [ ] **D.2 Fairness across groups**: Have we tested model results for fairness with respect to different affected groups (e.g., tested for disparate error rates)?
 - [ ] **D.3 Metric selection**: Have we considered the effects of optimizing for our defined metrics and considered additional metrics?

> Standard "accuracy" is going to be misleading here because the classes are imbalanced—most customers actually stay. If the model just guessed "No Churn" for everyone, it would still get a high accuracy score but be totally useless. We need to look at metrics like Recall (to make sure we actually catch the people leaving) or Precision instead, so we aren't just optimizing for a high score that doesn't solve the business problem.

 - [ ] **D.4 Explainability**: Can we explain in understandable terms a decision the model made in cases where a justification is needed?
 - [ ] **D.5 Communicate limitations**: Have we communicated the shortcomings, limitations, and biases of the model to relevant stakeholders in ways that can be generally understood?

> A big limitation is that we are predicting the future based entirely on the past. The data captures why people left historically, but it doesn't account for new things happening right now, like a new competitor showing up or a price hike. We need to make sure stakeholders know that the model assumes customer behavior will stay the same, which might not be true if the market changes.
